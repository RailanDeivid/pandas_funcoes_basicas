# Identificando valores nulos

A manipulação de valores ausentes é uma etapa crítica para a construção de uma boa análise de dados. Os valores ausentes devem ser uma prioridade máxima, pois têm um efeito significativo na precisão de qualquer análise.

### isnull()

* Essa função por si só retrona a mesma tabela, porém, onde os dados foram nulos true e false onde não forem.

```python
df.isnull()
```

### isna().any()

A função Isna seguida da função any() retorna as colunas onde contém valores nulos.

```python
df.isna().any()
```

### Verificando a soma de valores ausentes

Calculando o número total de valores ausentes na base de dados.
* Você pode calcular o número de valores ausentes em cada coluna por `df.isnull().sum()`

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
nan_cols = [i for i in df.columns if df[i].isnull().sum() >= 1]
for i in nan_cols:
    import numpy as np
    print(f"{i}:", "contém", np.round(df[i].isnull().mean() * 100, 2),
          '% dos dados nulos', 'e um total de',
          np.round(df[i].isnull().sum(), 4,), "linhas afetadas")
```

```python
print('{} Linhas e {} colunas'.format(df.shape[0], df.shape[1]))
print('Quantidade de dados nulos:', df.isna().sum().sum())
```

# Seleções em DataFrames

### Loc and iloc

* São usaddos para fazer seleções em linhas e colunas.


* **loc:** seleções por labels
* **iloc:** seleções por posições

#### Loc

loc é usado para selecionar dados pelo labels (rótulos). Os rótulos das colunas são os nomes das colunas.

* Os rótulos das linhas se iniciam em 0.

<h>

* Selecionando as 10 primeiras linhas e de duas colunas específicas

```python
df.loc[:9, ['Tipo', 'Valor']]
```

#### Loc com mais de uma condição.

* Para isso basta usar ‘&’ entre os filtros que devem ser colocados entre parênteses.

Selecionando dados onde o `Tipo`é apartamento e o `IPTU`seja menor ou igual a 800

```python
df.loc[(df['Tipo']== 'Apartamento') & (df['IPTU'] <= 800)]
```

#### Loc de acordo com strings.

Junto ao loc também pode usar outras funções, nesse caso podemos usar a função `contains()` como uma condição.

* Como nesse caso o que estamos buscando é um texto (string) temos que informar isso usando a função `str`

```python
# selecionando dados onde a coluna 'Bairro' contém o nome 'São Francisco'

df.loc[df['Bairro'].str.contains('São Francisco')][:10]
```

#### iloc

O método iloc é bem mais simples e mais restrito, iloc Retorna a seleção a partir dos index numéricos.

* As posições de linha que são usadas com iloc também são números inteiros começando em 0.

```python
df.iloc[:9, [1,6]]
```

## Query

Do modo mais básico pode se dizer que o uma query no pandas funciona com um loc por trás. Porém sua sintaxe e feita em strings, que devem ser colocadas entre aspas simples e as suas condições em aspas duplas ou vice e versa.



Podendo utilizar os operadores lógicos:

- AND (&)
- OR (|)
- In

* Query com uma condição.

```python
df.query('Bairro == "Barra da Tijuca" ')
```

* Query com mais de uma condição podemos usar o `&` ou `and`

```python
df.query('Bairro == "Barra da Tijuca" & Tipo == "Apartamento"')
```

* Usando `OR`

```python
df.query('Area <80 or Tipo == "Quitinete"')
```

* Usando o método `In`


Basicamente o parâmetro In é uma busca especifica podendo ser informada a partir de uma array com as condições.

```python
df.query('Valor in [800, 1000]')
```

* Query a partir de uma variável.

**Nesse métodos usaremos o modo `f-strings`**

```python
# lista com condições a serem buscados
bairros = ['Bangu','Campinho', 'Higienópolis']

df.query(f'Bairro in {bairros}')
```

* Ou podemos adicionar um @ antes da variável.

```python
bairros = ['Tijuca']

df.query('Bairro in @bairros')
```

* Query em colunas com espaço.


**Quando uma coluna do df contém espaço ela deve ser coloca entre uma crase que é o acento indicado pelo símbolo [ ` ]**

```python
# reescrevendo a coluna
df.rename(columns={'Area': 'Area Construida'}, inplace=True)

df.query(' `Area Construida` < 80')
```

### map()

A função `map()` é bem limitada, podendo ser aplicada apenas a uma série.

> Como funciona a função?

* Como argumentos além de funções o `map()` pode receber tanto uma series pandas ou um dicionário. 
    - Quando a função é aplicada em uma series ela pega cada elemento da série, aplica a função e retorna uma series transformada com os argumentos utilizados.
    
 > **Resumindo é usado para substituir cada valor de uma Série por outro valor, que pode ser derivado de uma função, a dict ou a Series**

```python
# filtrando apenas tipo de casa e apartamento
df_ = df.query('Tipo == "Casa" or Tipo == "Apartamento"')

# criando dict com os valores a ser alterados
tipo_uso = {
    'Apartamento': 'Residencial',
    'Casa': 'Residencial'
}

# aplicando função e subscrevendo valores
tipo_uso = df_['Tipo'].map(tipo_uso)

# mostrando resultado
tipo_uso
```

### insert()

 Usado para inseriri coluna em um local especifico no DataFrame setando o parametro `loc=` informando o index do df.
 
 * **Gera um ValueError se a coluna já estiver contida no DataFrame, a menos que `allow_duplicates` esteja definido como True.**

```python
df_.insert(loc=1, column='Tipo Imovel', value= tipo_uso)

df_.head()
```

### apply()

A função `apply()` pode ser usado para Data Frames e series pandas, onde seu argumento sempre será uma função.

> Como funciona a função?
* Quando você usa o `apply()` em uma série ou em um data frame, a função pega cada elemento na série e aplica o argumento em cada um deles, e seu retorno pode ser tanto uma series pandas ou um data frame.

* **Incluindo o valor por metro quadrado**

```python
# criando uma função para calcular o valor por metro quadrado
valor_m2 = [lambda df: df["Valor"] / df["Area Construida"] if df["Area Construida"] != 0 else 0 ]

# aplicando função
df['valor_m2'] = df.apply(valor_m2, axis=1)
```

```python
df.head()
```

### cut()

Usado para segmentar e classificar valores de dados em compartimentos. 
* Esta função também é útil para passar de uma variável contínua para uma variável categórica. 
   - Por exemplo, cut pode converter idades em grupos de faixas etárias. Suporta binning em um número igual de bins, ou uma matriz pré-especificada de bins.

* Criando uma classificação para os valores dos imoveis.

`- Popular, Padrão e Alto Padrão`

```python
# Deifinindo a faixa de valores e rotulos a serem considerados

valor_minimo = df['Valor'].min()
valor_maximo = df['Valor'].max()

rotulos = ['Popular', 'Padrão', 'Alto Padrão']

# è preciso N pontos para obter N - 1 classes
classes = [valor_minimo, 100000, 200000, valor_maximo]


# Aplicando função
df['classe_valor'] = pd.cut(x=df["Valor"], bins=classes, labels=rotulos, include_lowest=True)

# visualizando resultado
df.value_counts('classe_valor')
```

```python
df.sample(5)
```

# Tabelas Dinâmicas

```python
# agrupando variaveis
grupamento = df[['Tipo', 'classe_valor', 'valor_m2']].groupby(by=['Tipo','classe_valor'])

# atribuindo a função
estatisticas = grupamento.agg(['min', 'mean', 'max']).round(2)

# visualizando
estatisticas
```

## droplevel()

Retornar Series/DataFrame com índice/níveis de coluna solicitados removidos.

```python
estatisticas = estatisticas.droplevel(level=0, axis='columns')

estatisticas
```

## reset_index()

Redefina o índice ou um nível dele.

Redefina o índice do DataFrame e use o padrão. Se o DataFrame tiver um MultiIndex, esse método poderá remover um ou mais níveis.

```python
estatisticas.reset_index(inplace=True)
estatisticas
```

## pivot

 Retornar DataFrame remodelado organizado por valores de índice/coluna fornecidos.

```python
estatisticas.pivot(
    index='Tipo',
    columns='classe_valor',
    values='mean'

)
```

## pivot_table()

Cria uma tabela dinâmica no estilo de planilha como um DataFrame.

Os níveis na tabela dinâmica serão armazenados em objetos MultiIndex (índices hierárquicos) no índice e colunas do DataFrame resultado.

```python
df.pivot_table(
    values='valor_m2', # valores a serem demostrados
    index='Tipo', # variável a ser agrupada
    columns='classe_valor', # colunas
    fill_value="-", # visuzalização dos valores nulos
    margins=True, # ativando a coluna total no final
    margins_name='Média Geral', # nomeando
    aggfunc=['mean'] # calculo a ser feito
)
```
